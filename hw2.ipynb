while迴圈
定義函數a(x),並呼叫兩次函數

````python
def a(x):
  result = 0
  while result < 10:
    result = result + x
    if result==5:
      return result
  return result


b = a(1)
c = a(2)
print(b)
print(c)
````

1定義了一個遞迴函數 a(n)
2該函數計算從 1 到 n 的所有整數的總和

````python
def a(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n + a(n-1)

print(a(3))
````
